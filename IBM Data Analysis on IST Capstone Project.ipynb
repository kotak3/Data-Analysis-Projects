```python
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing the data
# Replace 'df' with your actual DataFrame name

# Get the counts of each database desired to learn next year
database_desire_counts = df['DatabaseDesireNextYear'].str.split(';', expand=True).stack().value_counts()

# Get the top 5 databases
top_5_databases = database_desire_counts.head(5)

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_5_databases, labels=top_5_databases.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.show()

# Calculate the percentage of MongoDB
mongodb_percentage = (top_5_databases['MongoDB'] / top_5_databases.sum()) * 100
print(f"The percentage of MongoDB among the top 5 databases respondents wish to learn next year is: {mongodb_percentage:.2f}%")
```

```python
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing the data
# Replace 'df' with your actual DataFrame name

# Filter the data for the age group 25-30
age_group_data = df[(df['Age'] >= 25) & (df['Age'] <= 30)]

# Group the data by age and calculate the median ConvertedComp for each age
median_comp_by_age = age_group_data.groupby('Age')['ConvertedComp'].median()

# Create the line chart
plt.figure(figsize=(10, 6))
plt.plot(median_comp_by_age.index, median_comp_by_age.values, marker='o', linestyle='-')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.title('Median ConvertedComp for Age Group 25-30')
plt.grid(True)
plt.show()
```
