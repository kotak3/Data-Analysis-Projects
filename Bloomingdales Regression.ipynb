```python

print(data_pl.columns.values)
data_pl['P4P'].describe()
```

```python
# Experimental Model with All Variables...
# Let's see what happens when we build a log-linear model with ALL columns of (ProfitLogic) data!
# This should be the best we can do, right?
#all_columns = "+".join(data_pl.columns - ['ln_R2'])
all_columns = "+".join(data_pl.columns[1:(len(data_pl.columns)-1)])
reg_formula = 'ln_R2~' + all_columns

fit_pl = smf.ols(formula = reg_formula, data = data_pl).fit()
print(fit_pl.summary())
```

```python
# What conclusions do you draw?  Is the model "stable"?  How about our variables of interest?
# Multi-collinearity is often a risk when using a lot of variables in a single regression...
# Let's try using only the variables of interest...(not in presentation)

some_columns = "+".join(data_pl.columns[1:5])
reg_formula2 = 'ln_R2~'+ some_columns
fit_pl2 = smf.ols(formula = reg_formula2, data = data_pl).fit()
print(fit_pl2.summary())
```
